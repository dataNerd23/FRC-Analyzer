# Part 2: Formulating $X\beta = Y$
Formulating the $X$ matrix from the equation $X\beta = Y$ where $X$ is the teams matrix (e.g., the total score of an alliance can be shown as frc2403+frc3853+frc498=Total Score), $\beta$ is a vector of the "averages"/best guess for each teams' corresponding score breakdown, and $Y$ is a vector of each team's match score breakdown (e.g., autoOwnershipPoints).  $\beta$ will be estimated via the method of least-squares ($\hat{\beta}=(X^TX)^{-1}X^TY$).


In addition, there are variables that are robot specific (i.e., blue.autoRobot1).  Since these variables already point to a specific robot, they are not necessary for the calculation $\hat{\beta}=(X^TX)^{-1}X^TY$.  Therefore, they will be added to $\hat{\beta}$ after the calculation.

```python
#Setting up the X matrix for the teams using Octave (Oct2Py)
import oct2py
oc = oct2py.Oct2Py()

xRows = 2*df.shape[0]
xCols = len(teamsTable)

#X = eng.zeros(xRows,xCols)
X = oc.zeros(xRows,xCols)

for xRowInd in range(xRows):
    if xRowInd%2 == 0:
        colShift = 0
        rowShift = xRowInd/2
    elif xRowInd%2 == 1:
        colShift = 3
        rowShift = (xRowInd-1)/2
    
    for dfColNum in range(0+colShift,3+colShift):
        colName = df.columns.values[dfColNum]
        indNum = teamsTable.index[teamsTable[0] == df[colName][rowShift]].tolist()[0]
        X[xRowInd][indNum] = 1
################################################################################################
```

```python
#Before setting up the Y matrix, data types need to be converted to numeric values.
#Specifically, boolean columns need to be converted to int64 and object columns to category

#Create array of robot specific stats (i.e., blue.autoRobot1)
robotSpecCols = [s for s in df.columns.values if "Robot" in s]
print(robotSpecCols)

for i in range(6,len(df.columns.values)):
    colName = df.columns.values[i]
    if df[colName].dtype == 'bool':
        df[colName] = df[colName].astype('int64')
    elif df[colName].dtype == 'object':
        df[colName] = df[colName].astype('category')

#Creating a numeric dataframe with only a subset of the data (exclude teams numbers, 
#just match data).  Also, removing robot specific stats (these stats would not be 
#helpful for Y matrix)
numMatchData = df
numMatchData = numMatchData.drop(robotSpecCols, axis=1)
numMatchData = pd.get_dummies(numMatchData.iloc[:, 6:len(numMatchData.columns.values)])
#test = numMatchData.columns.tolist()
#print(np.sort(test))
numMatchData = numMatchData[np.sort(numMatchData.columns.tolist())]

numMatchData.info()
numMatchData.head(10)
################################################################################################
```

```python
#Setting up the Y matrix for the teams using Octave (Oct2Py)
yRows = 2*numMatchData.shape[0]

#Find the center column index between blue and red alliance columns
nCol = len(numMatchData.columns.values)
evenCols = True
if nCol%2 == 0:
    yCols = int(nCol/2)
else:
    print('Number of columns is not even number: Please investigate data')
    evenCols = False

if evenCols:
    #Data in terms of offense perspective (i.e., Offensive Power Rankings (OPR))
    Y_Off = oc.zeros(yRows,yCols)
    
    #Data in terms of defensive perspective (i.e., Defensive Power Rankings (DPR))
    Y_Def = oc.zeros(yRows,yCols)

    for yRowInd in range(yRows):
        if yRowInd%2 == 0:
            colShift = 0
            rowShift = yRowInd/2
        elif yRowInd%2 == 1:
            colShift = yCols
            rowShift = (yRowInd-1)/2

        for dfColNum in range(yCols):
            colNameOff = numMatchData.columns.values[dfColNum+colShift]
            colNameDef = numMatchData.columns.values[dfColNum+(yCols-colShift)]
            
            Y_Off[yRowInd][int(dfColNum-colShift)] = numMatchData[colNameOff][rowShift]
            Y_Def[yRowInd][int(dfColNum-colShift)] = numMatchData[colNameDef][rowShift]

################################################################################################
```

```python
#Storing X and Y matrices in Octave, then perform least-squares on data to find beta hat
#Dimensions of X and Y matrices
print('Dimensions of X: '+str(X.shape))
print('Dimensions of Y_Off: '+str(Y_Off.shape))
print('Dimensions of Y_Def: '+str(Y_Def.shape))

#Store matrices in Octave
oc.push('X',X)
oc.push('Y_Off',Y_Off)
oc.push('Y_Def',Y_Def)

#Perform least-squares method
oc.eval("beta_Off = inv(X.'*X)*X.'*Y_Off;")
oc.eval("beta_Def = inv(X.'*X)*X.'*Y_Def;")

#Store matrices back into Python
beta_Off = oc.pull('beta_Off')
beta_Def = oc.pull('beta_Def')

#Dimensions of beta matrices
print('Dimensions of beta_Off: '+str(beta_Off.shape))
print('Dimensions of beta_Def: '+str(beta_Def.shape))
################################################################################################
```

```python
#Adding the robot specific attribtues to each robot

#Make empty dataframe to add robot specific data
roboSpecStats = pd.DataFrame(index=teamsTable[0], columns=[0])

for i in range(len(robotSpecCols)):
    #Getting the robot column name and general column name that corresponds that attribute
    #(i.e., blue1 and autoRobot from blue.autoRobot1)
    stringColSplit = robotSpecCols[i].split('.')
    alliColor = stringColSplit[0]
    robotAttriStrList = list(stringColSplit[1])
    
    robotColName = alliColor+str(robotAttriStrList[len(robotAttriStrList)-1])
    genRobotAttribute = ''.join(robotAttriStrList[0:(len(robotAttriStrList)-1)])
    
    #Get the unique value from each robot specific column
    uniqueAttriList = df[robotSpecCols[i]].unique()
    for j in range(len(uniqueAttriList)):
        #Returing the unique column names that result from the unique attribute
        newColName = genRobotAttribute+'.'+uniqueAttriList[j]
        
        #For first instance, add new column to roboSpecStats.  If the column already exists
        #in roboSpecStats, continue.  Else, add column to roboSpecStats
        if 0 in roboSpecStats.columns.values:
            roboSpecStats.rename(columns = {0: newColName}, inplace=True)
            roboSpecStats[newColName] = 0
        elif not (newColName in roboSpecStats.columns.values):
            roboSpecStats[newColName] = 0
            
        for k in range(len(teamsTable)):
            #Find the rows where the robot name appears in that specific column (i.e., "blue1")
            rowsTeams = df[robotColName][df[robotColName]==teamsTable.iloc[k][0]]
            
            #Sum the instances in which that team performed that task and add it to the previous
            #value in robotSpecCols (i.e., autoRobot.None = did not perform autonomous).
            #This is accomplished by exploiting the fact that "True" returns 1.
            value = (df.iloc[rowsTeams.index][robotSpecCols[i]] == uniqueAttriList[j]).sum()
            roboSpecStats[newColName][teamsTable.iloc[k][0]] = value + \
            roboSpecStats[newColName][teamsTable.iloc[k][0]]
roboSpecStats.head(60)
#print(roboSpecStats.shape)
################################################################################################
```
